```python
import yfinance as yf
import numpy as np
```

```python
#dados = formular_dados.acoes_ibov
dados = ['ABEV3.SA', 'EGIE3.SA', 'RECV3.SA', 'TUPY3.SA', 'VALE3.SA', 'BCFF11.SA', 'HGLG11.SA', 'TAEE4.SA', 'VIVT3.SA', 'WEGE3.SA', 'RADL3.SA']
acoes = sorted(dados)
display(acoes)
```

```python
# Obtendo os dados das ações
dados_acoes = yf.download(acoes, period = 'max', interval = '1mo')['Adj Close']
display(dados_acoes)
```

```python
# Calculando os retornos
dados_retornos = dados_acoes.pct_change()
display(dados_retornos)
```

```python
# Calculando a matriz de covariância
covariance_matrix = dados_retornos.cov()
display(covariance_matrix)
```

```python
# Calculando a media dos retornos
media_retornos = dados_retornos.mean()
display(media_retornos)
```

```python
# tornando as matrizes em array
matriz_cov = np.array(covariance_matrix)
retornos = np.array(media_retornos)
display(matriz_cov,retornos)
```

```python
# transformando a matriz covariancia na matra A da Função para o metodo de newton
matriz = np.empty([len(acoes),len(acoes)])

for i in range(len(acoes)):
    matriz[i] = (matriz_cov[i]*2)

# Adicionando uma coluna de uns à direita da matriz para se tornar A
nova_coluna = np.ones((matriz.shape[0], 1))
matriz_a = np.hstack((matriz, nova_coluna))

# Adicionando uma linha de uns e uma linha de zeros na parte inferior da matriz expandida para se tornar A
nova_linha_1 = np.ones((1, matriz_a.shape[1]))
matriz_a = np.vstack((matriz_a, nova_linha_1))

# Matriz A finalizada
matriz_a[len(acoes),len(acoes)] = 0
display(matriz_a)
```

```python

# Frunçõ que calcula o gradiente da função de otimização
def gradiente(p, matriz):
    grad = np.dot(p, matriz)
    grad[len(acoes)] = grad[len(acoes)]-1
    return grad
```

```python
# Definindo a função do gradiente que irá otimizar a carteira
def newton_raphson(p, q):
    d = gradiente(p,q)
    d = np.array(d)
    d = d*-1
    k = 0
    pant = d.copy()
    p = p + np.linalg.solve(q, d)
    while np.linalg.norm(p - pant) / np.linalg.norm(p) > 1e-10:
        pant = np.array(p)
        k += 1
        p = p + np.linalg.solve(q, d)
        d = gradiente(p,q)
        d = np.array(d)
        d = d*-1
        if k == 1000:
            break
    print(f'Executado {k} vezes')
    return p
```

```python
# Definindo a função do gradiente que irá otimizar a carteira
def newton_raphson_test(p, q):
    d = gradiente(p,q)
    d = np.array(d)
    d = d*-1
    k = 0
    pant = d.copy()
    inv = np.linalg.inv(q)
    p = p + np.dot(d,inv)
    while np.linalg.norm(p - pant) / np.linalg.norm(p) > 1e-10:
        pant = np.array(p)
        k += 1
        p = p + np.dot(d,inv)
        d = gradiente(p,q)
        d = np.array(d)
        d = d*-1
        if k == 1000:
            break
    print(f'Executado {k} vezes')
    return p
```

```python
# Definindo o ponto inicial para o gradiente
ponto = np.zeros(len(acoes))
ponto_zero = np.append(ponto, 1)
display(ponto_zero)
```

```python
# Minimizando a carteira
carteira_otimizada = newton_raphson(ponto_zero,matriz_a)
display(carteira_otimizada)
```

```python
# Função que calcula o risco a qual foi minimizada
def funcao_f(p):
    return (np.dot(np.dot(p, matriz_cov), p))
```

```python
carteira_minimizada = np.zeros(len(acoes))

for i in range(len(acoes)):
    carteira_minimizada[i] = carteira_otimizada[i]

i=0
for acao in acoes:
    print(f'{acao}:', f'Alocar {carteira_minimizada[i]*100:,.0f}% do seu capital')
    i += 1
```

```python
retorno_esperado = np.dot(carteira_minimizada, retornos)
risco_carteira = funcao_f(carteira_minimizada)

print(f'Esse é o retorno esperado para sua carteira: {retorno_esperado*100}%')
print(f'O risco mais baixo encontrado para sua carteira é: {risco_carteira*100}%')
```
